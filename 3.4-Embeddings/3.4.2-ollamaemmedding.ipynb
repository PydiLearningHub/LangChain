### Ollama

## Ollama supports embedding models, making it possible to build retrieval argumented generation(RAG) applciations that combine text prompts with existing documents or other data.

```python
from langchain_ollama import embeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma2:2b")  ##by default it uses llama2
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "Aplha is the first letter of Greek alphabet",
        "Beta is the second letter of Greek alphabet"
    ]
)
```

```python
r1
```

```python
embeddings.embed_query("What is the second letter of Greek alphabet ")
```
