#### Sentence Transformers on Hugging Face
Huggin Face sentence-transformers is a Python framework for state-of-the-art sentence, text and image embeddings. One of the embedding models is used in the HuggingFaceEmbeddings class. We have also added an alias for SentenceTransformerEmbeddings for users who are more familar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="this is a test documents"
query_result=embeddings.embed_query(text)
```

```python
query_result
```

```python
len(query_result)
```

```python
doc_results = embeddings.embed_documents([text, "This is not a test document."])
doc_results
```
